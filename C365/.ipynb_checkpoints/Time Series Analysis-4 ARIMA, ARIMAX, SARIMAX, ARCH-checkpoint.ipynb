# ARIMA Model

ARIMA = Auto Regressive Integrated Moving Averages Model

ARIMA has 3 parameters (p,d,q)
p = parameter of Auto Regressive
q = parameter of Moving Averages
d = The number of times we need to integrate the time series to ensure stationarity

ARIMA(p,0,q) = ARMA(p,q)
ARIMA(0,0,q) = MA(q)
ARIMA(p,0,0) = AR(p)

Intution: We are accounting for the non-seasonal difference between periods.
The AR components will be the change in prices and not the actual prices.
We will construct a new time series (d) and fit that in the ARMA(p,q) model.

## LLR Test

```python
def LLR_test(mod_1, mod_2, DF = 1):
    L1 = mod_1.fit(start_ar_lags = 15).llf
    L2 = mod_2.fit(start_ar_lags = 15).llf
    LR = (2*(L2-L1))    
    p = ssd.chi2.sf(LR, DF).round(3)
    return p
```

```python
df['returns'] = df.market_value.pct_change(1)*100
df = df.iloc[1:]
```

```python
model_ar_1_i_1_ma_1 = sta.ARIMA(df.market_value, order = (1,1,1))
result_ar_1_i_1_ma_1 = model_ar_1_i_1_ma_1.fit()
result_ar_1_i_1_ma_1.summary()
```

```python
df['res_ar_1_i_1_ma_1'] = result_ar_1_i_1_ma_1.resid
sgt.plot_acf(df.res_ar_1_i_1_ma_1[1:], zero = False, lags = 40)
plt.title("ACF of residuals of ARIMA(1,1,1)", size = 20)
plt.show()
```

## SARIMAX

SARIMAX stands for seasonal ARIMAX. Here we take into account the seasonality of the data. For example if there are jump in the views of jingle bells video on youtube in the month of december, this has to be taken into account as the seasonality parameter. If a model is fed with only the december data it will predict the same kind of outcomes in the month of January or even the next month november which is terribly wrong. To overcome this problem we go for SARIMAX models. For more details please check the supplementary pdf in the pdf folder.

```python
Image(filename='SARIMAXEqn.jpg')
```

The lower case letter (phi and theta) are non seasonal factors while the upper case letters are seasonal ones.

## ARCH

AutoRegressive Conditional Heteroskedasticity (Heteroscedasticity) Model. Heteroscedasticity means "Different Dispersion". The most commom versions of dispersion are variance (sigma square) and standard deviation (sigma). Since we have direct relationship between the two, we are only going to discuss the variance. In bayesian inference the term conditional means "A value that depends upon others". The simplest example can be conditional probablity in this case where the likelyhood of one outcome varies when conditioned to other outcomes. So in this model we will be dealing with variance conditioned on other values, these other values will be coming from the AutoRegressive model.

Suppose we have a variable y then its ARCH model equation will be:

We use ARCH models to measure volitility. Volitility refers to numeric measurement of uncertainity which depends upon the volume of changes in valuation. Valuation here refers to prices or returns of the said portfolios or securities or market indices. Volatility is not directly observable. In our project we are not monitoring data throughout the day we are only dealing with the closing prices of the day so the intra day volitility cannot be measured. Volitlity evolves continuosly. Volitility never goes to infinity. 

"EGARCH" is a kind of model you should look up online. Its based on the volitility of returns.

Investors prefers stability. If 2 portfolios provides equal returns but one is highly volatile among them the investors will move to the more stable option.

ARCH has mean and variance equations. 

    r(t) = mean(t) + E(t)
    
    mean(t) = A function of past values or past errors.
    E(t)    = The residual values left after estimating the coefficients.
    
We assume that mean is serially uncorrelated i.e. there is no time-dependent pattern.

### The Simple ARCH model

```python
model_arch_1 = arch_model(df.returns[1:], mean = 'constant', vol = 'ARCH', p = 1)
results_arch_1 = model_arch_1.fit()
results_arch_1.summary()
```

Reading the results:

1. Our model took 6 iterations, so its a less resource consuming model and we can expect quick results.
2. R-squared and adjusted R-squared both are negative zero. There is nothing called as negative zero, it means the value is negative but so minute that it can be considered as zero.
3. R-squared is the measurement of explanatory variation away from the mean.
4. If the residuals are simply a version of the original dataset, where every value is decreased by a constant, then there will be no actual variance. There is nothing to explain in such case of residuals and hence R-squared is zero.
5. R-squared is a useful measurement for other models but not the ARCH models.
6. If you check the Log-likelyhood value of the model, you will see that when we compare it to the ARMA and ARIMA models its value is very high. So, Even the simplest ARCH model yields a better estimate than the complex ARIMA models. Which is great as the higher complexity is immediately paying off.
7. The DF-model whose value is 3 is nothing but the coefficients determined by the model.
8. In the mean model, the p-value is very low. We read it as if the value after the e is negative than the number is getting very small if it has a positive value than the number is getting really high. We also see that the coefficient of the mean variable is significant and this will be our baseline for residuals.
9. Volitility model also know as the need from the ARCH model. We see that both alpha and omega are significant and they are constant terms for the variance equation. Also the p-values of both the constants are really small.
10. ARCH model has instantly become a better estimator than the complex ARIMA models we have worked on so far.

However, ARCH can only be used for prediction of future variance rather than future returns. Thus, we can use it to check the stability in the market but not for predicting about the price's ups and downs.
